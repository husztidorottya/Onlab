```python
import sys
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import nltk
from wordcloud import WordCloud, STOPWORDS

# set plot properties
matplotlib.style.use('ggplot')
matplotlib.pyplot.rcParams['figure.figsize'] = (25, 13)
matplotlib.pyplot.rcParams['font.family'] = 'sans-serif'
matplotlib.pyplot.rcParams['font.size'] = 20

# read the experiment result
df = pd.read_csv('log_experiment.csv')

# visualize the score according to the average of embeddingDim
df.groupby(by=[df.embedding]).mean().plot(y='score', kind='bar', color='blue')

# show table grouped by embeddingDim
columns = ['score']
df_embedding = df[columns]
df_embedding.groupby(by=[df.embedding]).mean()
```

```python
# visualize the score according to the average of LSTM cells
df.groupby(by=[df.LSTM]).mean().plot(y='score', kind='bar')

# show table grouped by LSTM cells
df_LSTM = df[columns]
df_LSTM.groupby(by=[df.LSTM]).mean()
```

```python
# visualize the score according to the average of batch size
df.groupby(by=[df.batch]).mean().plot(y='score', kind='bar', color='green')

# show table grouped by batch size
df_batch = df[columns]
df_batch.groupby(by=[df.batch]).mean()
```
